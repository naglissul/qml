## Bell states

## $\ket{\Phi^+}$

```python
from qiskit import QuantumCircuit

# Basic entanglement diagram (Bell state \ket{\Phi^+})
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.draw('mpl')
```

## $\ket{\Phi^-}$

```python
# Basic entanglement diagram (Bell state \ket{\Phi^-})
qc = QuantumCircuit(2)
qc.x(0)
qc.h(0)
qc.cx(0, 1)
qc.draw('mpl')
```

```python
qc2 = QuantumCircuit(2)
qc2.h(0)
qc2.cx(0, 1)
qc2.draw('mpl')
qc2.z(0)
qc2.draw('mpl')
```

## $\ket{\Psi^-}$

```python
# Basic entanglement diagram (Bell state \ket{\Psi^-})
qc = QuantumCircuit(2)
qc.x(1)
qc.x(0)
qc.h(0)
qc.cx(0, 1)
qc.draw('mpl')
```

```python
from qiskit.circuit.library import UnitaryGate
import numpy as np

iY_matrix = np.array([[0, -1j], [1j, 0]])
iY_gate = UnitaryGate(iY_matrix, label='iY')

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.append(iY_gate, [0])

qc.draw('mpl')
```

# Let's try at least the phase kickback


```python
qc = QuantumCircuit(2)

qc.h(0)         # Apply Hadamard to qubit 0
qc.y(1)
qc.cy(0, 1)     # Apply controlled-Y (CY) gate with qubit 0 as control and qubit 1 as target

qc.measure_all()  # Measure all qubits
qc.draw('mpl')
```
